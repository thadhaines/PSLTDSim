```python

gradLin = []
for clr in range(10):
    sClr = (clr)/10
    gradLin.append([sClr, sClr, sClr])
    
print(gradLin)
```

```python
gradExp = []
for clr in range(7):
    sClr = (1-np.exp(-(clr)/7.5))
    print(sClr)
    gradExp.append([sClr, 1-sClr, sClr])
    
print(gradExp)
```

```python
lines = ['-',  (0, (5, 1)),(0, (1, 1)),]
lwidth = [1,1.2,1.4]
```

```python
from cycler import cycler
cc = (cycler(linestyle=lines)+cycler(linewidth=lwidth))*cycler(color = gradLin) # *cycler(linewidth=lwidth))
#for d in cc:
#    print(d)
```

```python
x = np.arange(1,np.pi*4,20/1000)
y = np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.set_prop_cycle(cc)
a = range(len(cc))
for num in a:
    #ax.plot(x,np.sin(x+num/np.pi)*num, label=str(num))
    ax.plot(x,np.sin(x)*10-num, label=str(num+1))
ax.legend(loc=5)
ax.grid(True)
fig.set_dpi(150)

fig.set_size_inches(9, 4)
plt.show()
```
