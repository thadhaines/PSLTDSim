```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
```

```python

name = "tab10"
cmap = get_cmap(name)
print(cmap)
colors = cmap.colors
#print(colors)
cDark= np.array(colors)*(.5,.5,.5)
#print(cDark)
customClrs = [
    [0,.5,1],  # blue
    [1, .5, 0],  # orange
    [1, 0, 1], # magenta
    [0, 1, 0], # green  
    [0, 1, 1], # cyan
    
    [.6, .6, .6], # grey  
    #[0, 0, 0], # black
    [0, 0.2, 0], # dark green
    [1, 0, 0], # red
]
```

```python
lines = ['-',  (0, (5, 1)),(0, (1, 1)),]
lwidth = [1,1.2,1.4]
```

```python
from cycler import cycler
cc = (cycler(linestyle=lines)+cycler(linewidth=lwidth))*cycler(color = customClrs) # *cycler(linewidth=lwidth))
#for d in cc:
#    print(d)
```

```python
x = np.arange(1,np.pi*4,20/1000)
y = np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.set_prop_cycle(cc)
a = range(len(cc))
for num in a:
    #ax.plot(x,np.sin(x+num/np.pi)*num, label=str(num))
    ax.plot(x,np.sin(x)*10-num, label=str(num+1))
ax.legend(loc=5)
ax.grid(True)
fig.set_dpi(150)

fig.set_size_inches(9, 4)
plt.show()
```
